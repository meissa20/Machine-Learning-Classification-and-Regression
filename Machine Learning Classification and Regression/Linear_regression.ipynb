```python
import pandas as pd
```

```python
Houses = pd.read_csv("California_Houses.csv")
```

```python
Houses
```

```python
Houses[Houses.isnull().any(axis=1)]
```

```python
Houses.corr()
```

```python
from sklearn.model_selection import train_test_split

# Split into 70% training and 30% temporary (validation + test)
train_set, temp_set = train_test_split(Houses, test_size=0.3, random_state=10)

# Split temp_set into 15% validation and 15% test
val_set, test_set = train_test_split(temp_set, test_size=0.5, random_state=10)

print(f"Train set: {len(train_set)} rows")
print(f"Validation set: {len(val_set)} rows")
print(f"Test set: {len(test_set)} rows")
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

reg_model = LinearRegression()
ridge_model = Ridge(alpha=1.0)
lasso_model = Lasso(alpha=1.0, max_iter=5000)
```

```python
predictors = ["Median_Income", "Median_Age", "Tot_Rooms", "Tot_Bedrooms", "Population", "Households", "Latitude", "Longitude", "Distance_to_coast", "Distance_to_LA", "Distance_to_SanDiego", "Distance_to_SanJose", "Distance_to_SanFrancisco"]
```

```python
reg_model.fit(train_set[predictors], train_set["Median_House_Value"])
```

```python
ridge_model.fit(train_set[predictors], train_set["Median_House_Value"])
```

```python
lasso_model.fit(train_set[predictors], train_set["Median_House_Value"])
```

```python
val_predictions_reg_model = reg_model.predict(val_set[predictors])
val_predictions_ridge_model = ridge_model.predict(val_set[predictors])
val_predictions_lasso_model = lasso_model.predict(val_set[predictors])
```

```python
val_predictions_reg_model = val_predictions_reg_model.round()
val_predictions_ridge_model = val_predictions_ridge_model.round()
val_predictions_lasso_model = val_predictions_lasso_model.round()

val_set["predictions_reg_model"] = val_predictions_reg_model
val_set["predictions_ridge_model"] = val_predictions_ridge_model
val_set["predictions_lasso_model"] = val_predictions_lasso_model
```

```python
test_predictions_reg_model = reg_model.predict(test_set[predictors])
test_predictions_ridge_model = ridge_model.predict(test_set[predictors])
test_predictions_lasso_model = lasso_model.predict(test_set[predictors])
```

```python
test_predictions_reg_model = test_predictions_reg_model.round()
test_predictions_ridge_model = test_predictions_ridge_model.round()
test_predictions_lasso_model = test_predictions_lasso_model.round()

test_set["predictions_reg_model"] = test_predictions_reg_model
test_set["predictions_ridge_model"] = test_predictions_ridge_model
test_set["predictions_lasso_model"] = test_predictions_lasso_model
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_val_reg_model = mean_absolute_error(val_set["Median_House_Value"], val_set["predictions_reg_model"])
mae_val_ridge_model = mean_absolute_error(val_set["Median_House_Value"], val_set["predictions_ridge_model"])
mae_val_lasso_model = mean_absolute_error(val_set["Median_House_Value"], val_set["predictions_lasso_model"])

mse_val_reg_model = mean_squared_error(val_set["Median_House_Value"], val_set["predictions_reg_model"])
mse_val_ridge_model = mean_squared_error(val_set["Median_House_Value"], val_set["predictions_ridge_model"])
mse_val_lasso_model = mean_squared_error(val_set["Median_House_Value"], val_set["predictions_lasso_model"])

##############################################################################################################

mae_test_reg_model = mean_absolute_error(test_set["Median_House_Value"], test_set["predictions_reg_model"])
mae_test_ridge_model = mean_absolute_error(test_set["Median_House_Value"], test_set["predictions_ridge_model"])
mae_test_lasso_model = mean_absolute_error(test_set["Median_House_Value"], test_set["predictions_lasso_model"])

mse_test_reg_model = mean_squared_error(test_set["Median_House_Value"], test_set["predictions_reg_model"])
mse_test_ridge_model = mean_squared_error(test_set["Median_House_Value"], test_set["predictions_ridge_model"])
mse_test_lasso_model = mean_squared_error(test_set["Median_House_Value"], test_set["predictions_lasso_model"])
```

```python
print("mae_val_reg_model", mae_val_reg_model)
print("mae_val_ridge_model", mae_val_ridge_model)
print("mae_val_lasso_model", mae_val_lasso_model)

print("\nmse_val_reg_model", mse_val_reg_model)
print("mse_val_ridge_model", mse_val_ridge_model)
print("mse_val_lasso_model", mse_val_lasso_model)

##############################################################################################################

print("\n\n\nmae_test_reg_model", mae_test_reg_model)
print("mae_test_ridge_model", mae_test_ridge_model)
print("mae_test_lasso_model", mae_test_lasso_model)

print("\nmse_test_reg_model", mse_test_reg_model)
print("mse_test_ridge_model", mse_test_ridge_model)
print("mse_test_lasso_model", mse_test_lasso_model)
```
